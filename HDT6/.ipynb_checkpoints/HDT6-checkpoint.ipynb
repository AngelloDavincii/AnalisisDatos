# EJERCICIO 2

```python
data = pd.read_csv("datos.csv")
data
```

```python
##Espana, Italia, EEUU, GUATEMALA
#Extraemos informacion de la fila donde coincide el nombre con el pais que buscamos
Esp = data.loc[data["Country/Region"]=="Spain"]
Ita = data.loc[data["Country/Region"] == "Italy" ]
USA = data.loc[data["Country/Region"] == "US" ]
Gua = data.loc[data["Country/Region"] == "Guatemala" ]
```

```python
#Creamos datos para el eje x que son los dias 
x = []
for i in range(0,63):
    x.append(i)
```

```python
#Nos devuelve un ndarray con los valores de la fila del pais
w1 = Esp.values
w2 = Ita.values
w3 = USA.values
w4 = Gua.values
```

```python
#eliminamos las primeras 4 columnas de cada fila
y1=np.delete(w1,slice(4),None)
y2=np.delete(w2,slice(4),None)
y3=np.delete(w3,slice(4),None)
y4=np.delete(w4,slice(4),None)
```

```python
#Graficamos para cada pais
#Lo pude haber hecho funcion pero queria practicar
plt.style.use("ggplot")
plt.plot(x,y1)
plt.title("Coronavirus en España")
plt.ylabel("Infectados")
plt.xlabel("Dias transcurridos")
```

```python
plt.plot(x,y2)
plt.title("Coronavirus en Italia")
plt.ylabel("Infectados")
plt.xlabel("Dias transcurridos")
```

```python
plt.plot(x,y3)
plt.title("Coronavirus en Estados Unidos")
plt.ylabel("Infectados")
plt.xlabel("Dias transcurridos")
```

```python
plt.plot(x,y4)
plt.title("Coronavirus en Guatemala")
plt.ylabel("Infectados")
plt.xlabel("Dias transcurridos")
```

```python
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score


def find(x,y1,title):
    #Definimos funciones a la que queremos que se ajuste
    #nuestros datos
    def fun(x,c1,c2):
        return c1 + c2**log(x)
    
    def fun2(x,c1,c2):
        return c1 + e**(c2*x)
    
    def fun3(x,c1,c2):
        return c1 + c2*x
    
    
    
    #en palabras de python
    #returns: Optimal values for the parameters so that the sum of the squared
    #Es decir un array con los coeficientes que hacen que se acerque la funcion
    #a los datos
    c,cov = curve_fit(fun,x,y1)
    c2,pcov = curve_fit(fun2,x,y1)
    c3,cov = curve_fit(fun3,x,y1)
    
    #Creamos listas donde agregamos los nuevos valores para "y"
    #usando la funcion con los nuevos coeficientes
    m = []
    n = []
    o = []
    
    for i in range(0,63):
        m.append(fun(i,c[0],c[1]))
        n.append(fun2(i,c2[0],c2[1]))
        o.append(fun3(i,c3[0],c3[1]))
    
    
    #Plot practice!!!!!
    plt.plot(x,y1,label = "Datos",color = "r")
    plt.plot(x,m,label = "log(x)",color = "y")
    plt.plot(x,n,label = "e(x)",color="g")
    plt.plot(x,o,label = "ax")
    
    
    plt.title("Coronavirus en "+title)
    plt.ylabel("Infectados")
    plt.xlabel("Dias transcurridos")
    plt.legend()
    
    print("El R^2 con log(x): ", r2_score(y1,m))
    print("El R^2 con e(x): ", r2_score(y1,n))
    print("El R^2 con ax: ", r2_score(y1,o))
find(x,y1,"Espana")
```

### Por lo tanto para Espana el modelo que mejor se ajusta es e(x)

```python
find(x,y2, "Italia")
```

### Por lo tanto para Italia el mejor modelo que se ajusta es e(x)

```python
find(x,y3, "Estados Unidos")
```

### Para estados Unidos el modelo que mejor se ajusta es el exponencial

```python
find(x,y4, "Guatemala")
```

### Para Guatemala el modelo que mejor se ajusta igualmente es el exponencial

```python
dat = pd.read_csv("data2.csv",delimiter=';',header=0)
dat
```

```python
x=dat["x"]
y = dat["y"]

plot(x,y)
```

```python
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score
#Use la misma funcion que utilice arriba solo le cambien la funcion con la
#que queria que se apegaran los datos y listo :)
def find(x,y1):
    def fun(x,c1,c2,c3):
        return c1 + sin(c2*x+c3)
    
    c,cov = curve_fit(fun,x,y1)

    m = []
    for i in range(0,33):
        m.append(fun(i,c[0],c[1],c[2]))   

    
    plt.plot(x,y1,label = "Datos",color = "r")
    plt.plot(x,m,label = "log(x)",color = "y")

    
    
    plt.title("Data vs A + sin(B x + C)")
    plt.ylabel("y")
    plt.xlabel("x")
    plt.legend()
    print("Los coeficientes de la formula son: ",c)
    print("El R^2 con log(x): ", r2_score(y1,m))
    
find(x,y)
```
