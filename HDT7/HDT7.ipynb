```python
from pylab import *
import pandas as pd
from matplotlib import pyplot as plt
import scipy.interpolate as si
import sympy as sy
```

```python
a, b, h = 0,2*pi,0.1 

x = arange(a,b,h) #generamos puntos de 0 a 2pi con un paso de 0.1
y = sin(x) #generamos datos con funcion original
primey = cos(x) #generamos datos con derivda
twoprimey = -sin(x) #generamos datos de sgunda derivada
```

```python
#Richardson definida en clase
def richardson(y,h):
    N = len(y)
    dy = zeros(N)
    for k in range(2,N-2):
        dy[k] = (4/3)*(y[k+1]-y[k-1])/(2*h)-(1/3)*(y[k+2]-y[k-2])/(4*h)
    return dy
```

```python
#aplicamos richardson a la primera y y luego al array de la primera derivda
#luego limiamos los array para que tengan la dimension adecuada
#debido a que perdemos datos al sacar derivdas

y1 = richardson(y,h)
y1 = y1[2:-2]
x1 = x[2:-2]
x2 = x[4:-4]
y2 = richardson(y1,h)
y2 = y2[2:-2]

plt.plot(x,y,label = "sin(x)")
plt.plot(x1,y1,label = "cos (x)")
plt.plot(x2,y2,label = "- sin (x)")

legend()
grid()
```
